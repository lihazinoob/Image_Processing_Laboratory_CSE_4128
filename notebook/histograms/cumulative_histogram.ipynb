```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("../../data/Lena.jpg",cv2.IMREAD_GRAYSCALE)
# Getting the height and width of the image
height,width = image.shape
histogram = np.zeros(256,dtype=np.uint8)
cumulative_histogram = np.zeros(256,dtype = np.uint16)
```

```python
# Manually iterating over the image matrix and calculate the cumulative histogram
for y in range(height):
  for x in range(width):
    # calculating the original histogram
    intensity = int(image[y,x])
    histogram[intensity] +=1

cumulative_histogram[0] = histogram[0]
for i in range(1,256):
  cumulative_histogram[i] = cumulative_histogram[i-1] + histogram[i]

print(cumulative_histogram)
```

```python
plt.figure(figsize=(8,5))
plt.title("Grayscale Histogram (Manual Calculation)")
plt.xlabel("Pixel Intensity (0-255)")
plt.ylabel("Number of Pixels")
plt.plot(cumulative_histogram, color='black')
plt.xlim([0, 256])
plt.show()
```
