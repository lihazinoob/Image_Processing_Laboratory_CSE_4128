```python
# This code is for creating a basic histogram and visualize it
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("../../data/Lena.jpg",cv2.IMREAD_GRAYSCALE)
# Getting the height and width of the image
height,width = image.shape
# First create an array of 8 bits or 256 entries
histogram = np.zeros(256,dtype = np.uint8)
print(histogram[0])
```

```python
# Manually iterating over the whole image and create a histogram
for y in range(height):
  for x in range(width):
    intensity = int(image[y,x])
    histogram[intensity]+=1

print(histogram)
    
```

```python
# Plotting the histogram
plt.figure(figsize=(8,5))
plt.title("Grayscale Histogram (Manual Calculation)")
plt.xlabel("Pixel Intensity (0-255)")
plt.ylabel("Number of Pixels")
plt.plot(histogram, color='black')
plt.xlim([0, 256])
plt.show()
```
