```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
origina_image = cv2.imread("../../data/histogram.jpg",cv2.IMREAD_GRAYSCALE)

original_image_histogram = cv2.calcHist([origina_image],[0],None,[256],[0,256])

equalized_image = cv2.equalizeHist(origina_image)

equalized_image_histogram = cv2.calcHist([equalized_image],[0],None,[256],[0,256])



```

```python
# Show the output images
plt.figure(figsize=(12,16))
plt.subplot(3,2,1)
plt.imshow(origina_image, cmap='gray')
plt.title('Original Image')
plt.axis('off')


plt.subplot(3,2,2)
plt.plot(original_image_histogram)
plt.title('Original Image Histogram')
plt.axis('on')

plt.subplot(3,2,3)
plt.imshow(equalized_image, cmap='gray')
plt.title('Equalized Image')
plt.axis('off')

plt.subplot(3,2,4)
plt.plot(equalized_image_histogram)
plt.title('Equalized Image Histogram')
plt.axis('on')
plt.tight_layout()
plt.show()
```
